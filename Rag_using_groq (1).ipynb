```python
from groq import Groq
```

```python
client=Groq(api_key="***************")
```

```python
docs=["I am haritha and i started learning machine learning .",
      "Machine learning is the basics of AI AND they consists of  large number of libraries ",
      "The library Numpy is used for mathetical data",
      "the library Pandas is used for dataframe",
      "The library Matplotlib and seaborn is used for Data visualization",
      "there are many other libraries which where used for anlyasis"]   # 6 documents because they were seperated by commas
```

```python
model = SentenceTransformer('sentence-transformers/all-MiniLM-L6-v2')
embedded=model.encode(docs).astype('float32') #here we want to embedded so we want to first encode and use the
```

```python
embedded.shape
```

```python
embedded
```

```python
dimension=embedded.shape[1] # (6,384 )  We are taking only 384
index=faiss.IndexFlatL2(dimension) # L2-contains  the eucledian distance to  finds the similarity, flat- save the input without any changes in dtype , index - same indx
index.add(embedded)  # here we are combining the index and embedded  to find the simalirty and store in the faiss
# here embedded has the oringinal vector
# index contains the eucledian distance
```

```python
# now we want to ask the ques
query = "Who learns machine learning?"
queryEmbedded=model.encode([query]).astype("float32") # because thellm can understand only the embedded text which in the vector dtype
top=2 # fetches teh top values of n , here n=2
dist, indices=index.search(queryEmbedded,top)  # dist finds the distance , indices tells about the document number where the answer peresent
retrived_chunks=[docs[i] for i in indices[0]]   # 2 for loops because the docs is 2D
#docs[i]  checks the documnet , indices[0]
print("Retrieved Chrunks.") # answer we find
for chunk in retrived_chunks:
  print("--",chunk)
```

```python
context="\n".join(retrived_chunks) # to combine all the documents into a single error
prompt=f"""
Answer the question based only on the context below.
Context:
{context}
Question:
{query}
Answer:
"""

response=client.chat.completions.create(    # it is used for getting the responses
    model='llama-3.3-70b-versatile',
    messages=[
              {"role":"user","content":prompt}],
              temperature=0.2
)
print("Final Answer : \n")
print(response.choices[0].message.content)
```
